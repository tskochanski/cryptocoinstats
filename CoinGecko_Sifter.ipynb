```python
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup
import re, os
import random
import time
import datetime
now = datetime.datetime.now()
dateStr = now.strftime('%Y-%m-%d')
print(dateStr)

url_0 = 'https://www.coingecko.com/en?page='
url_1 = 'https://www.coingecko.com/en/coins/'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
```

```python
pd.set_option('mode.chained_assignment', None)

url_A = pd.DataFrame()
x = range(1,2)
for n in x:
    r = requests.get(url_0 + str(n), headers=header)    
    dfs = pd.read_html(r.text)
    for df in dfs:
        df['CoinA']=''
        for m in range(len(df)):
            if (len(df.Coin[m].split()) < 4):
                df['CoinA'][m] = url_1 + df.Coin[m].split()[0]
            else:
                df['CoinA'][m] = url_1 + df.Coin[m].split()[0] + '-' + df.Coin[m].split()[1]
    url_A = url_A.append(df, ignore_index = True)
    time.sleep(random.randrange(2,6))
# print(url_A)
```

```python
url_A['happyness'] = ''
url_A['likers'] = ''
x = range(len(url_A))
# x = range(3)

for n in x:    
    r = requests.get(url_A.CoinA[n].lower(), headers=header)
    soup = BeautifulSoup(r.text, 'html.parser')
    progress = soup.findAll('div', {'class': 'progress-bar bg-success'} )
    stars = soup.findAll('span', {'data-controller': 'portfolios'} )
    prog_text = []
    stars_text = []
    for x in progress:
        prog_text.append(str(x))
    raw_txt=(str(prog_text))
    pp3=(raw_txt[1:len(raw_txt)-1]) 
    for x in stars:
        stars_text.append(str(x))
    raw_txt2=(str(stars_text))
    pp3B=(raw_txt2[1:len(raw_txt2)-1]) 
    
    
    try:
        url_A['happyness'].loc[n] = re.search(r'(?<=width:)(.*)(?=\"\>)', pp3).group(1).strip()       
        url_A['likers'].loc[n] = re.search(r'((\d+)|(\d+\,\d+))(?=\speople like this)', pp3B).group(1).strip().replace(',', '')
    except: 
        print('error - ',url_A['Coin'].loc[n],url_A['#'].loc[n] )    
    time.sleep(random.randrange(2,6))
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(url_A[['CoinA', 'happyness', 'likers']].sort_values('likers', ascending=True))
```
